```python
import gtts
import IPython.display as ipd
```

```python
def speak(text, lang='en', autoplay=True, echo=False):
    '''
    This will use gTTS to take your text and speak it.
    
    Example:
    
    speak("Hello!")
    '''
    speech = gtts.gTTS(text, slow=False, lang=lang)
    speech.save("demo.mp3")
    if echo:
        print(text)
    return ipd.Audio("demo.mp3",autoplay=autoplay)
```

```python
help(speak)
```

```python
speak("Yo mama!", lang="en", echo=True)
```

```python
text = "missippi"
search = "i"

def find(text,character):
    '''
    this will find character in text and if its there return True
    '''
    for ch in text:
        if ch == character:
            return True
        
    return False

print("when find('mississippi','i') expect True, Actual: ", find('mississippi', 'i') )
print("when find('mississippi','w') expect False, Actual: ", find('mississippi', 'w') )
```

```python
while True:
    ipd.display(speak("Enter some text ", echo =True))
    text = input()
    ipd.display(speak("Enter a single character", echo = True))
    ch = input()
    if find(text, ch):
        ipd.display(speak("I found it!", echo=True))
    else:
        ipd.display(speak("I did not find it!", echo=True))
        
    ipd.display(speak("Play Again???? ", echo=True))
    again = input()
    if again != 'y':
        break 
```
